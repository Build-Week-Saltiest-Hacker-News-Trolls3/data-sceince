```python
import time
import sqlite3
import urllib.request as rqst
from time import sleep
conna = sqlite3.connect('HackerRankTopDown.sqlite3')
ca = conna.cursor()
```

```python
for i in range(21500000,21568748):
    html = rqst.urlopen('https://hacker-news.firebaseio.com/v0/item/' + str(i) + '.json').read()
    ca.execute('INSERT INTO RawJSON (rawtxt, ID) VALUES (?,?)', [str(html), i])
    conna.commit()
    sleep(.01)
    if i%50 == 0:
        print(i)
    if i%1000 == 0:
        print('Commit 1000' + str(i))
        print(time.asctime())
```
