```python
import time
import sqlite3
import urllib.request as rqst
from time import sleep
connb = sqlite3.connect('HackerRankTopDown1.sqlite3')
cb = connb.cursor()
```

```python
for i in range(21496000, 21500000):
    html = rqst.urlopen('https://hacker-news.firebaseio.com/v0/item/' + str(i) + '.json').read()
    cb.execute('INSERT INTO RawJSON (rawtxt, ID) VALUES (?,?)', [str(html), i])
    connb.commit()
    sleep(.001)
    if i%50 == 0:
        print(i)
    if i%1000 == 0:
        print('Commit  ' + str(i))
        print(time.asctime())
```
