```python
import time
import sqlite3
import urllib.request as rqst
from time import sleep
connc = sqlite3.connect('HackerRankTopDown2.sqlite3')
cc = connc.cursor()


for i in range(21356000, 21400000):
    html = rqst.urlopen('https://hacker-news.firebaseio.com/v0/item/' + str(i) + '.json').read()
    cc.execute('INSERT INTO RawJSON (rawtxt, ID) VALUES (?,?)', [str(html), i])
    connc.commit()
    sleep(.001)
    if i%50 == 0:
        print(i)
    if i%1000 == 0:
        print('Commit  ' + str(i))
        print(time.asctime())
```
